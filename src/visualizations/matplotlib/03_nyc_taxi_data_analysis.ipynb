# NYC Taxi Data Analysis

```python
from matplotlib.lines import lineStyles
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.ion()

plt.get_backend()
plt.plot(np.random.randn(100).cumsum())
plt.show()
```

```python
df1 = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 1.csv")
df2 = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 2.csv")
df3 = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 3.csv")
df4 = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 4.csv")
df5 = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 5.csv")
df_full = pd.concat([df1,df2,df3,df4,df5])
df_full.head()
```

```python
df_full.info()
```

```python
df_full.describe()
```

```python
#change date format
df_full["Pickup_Date"] = pd.to_datetime(df_full["Pickup_Date"], format="%d/%m/%Y")
df_full.info()
```

```python
df_full.head()
trips = df_full.groupby(by='Pickup_Date', as_index=False)["ID"].count()
trips.head()
```

```python
trips.rename(columns={"ID": "Trips"}, inplace=True)
trips.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(trips["Pickup_Date"], trips["Trips"])
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Creating scatter plot for JFK Trip vs Tips data
taxi_trips = pd.read_csv("../../../data/matplotlib/Yellow Taxi Trips - Week 1.csv")
taxi_trips.head()
```

```python
JFK_Trips = taxi_trips[(taxi_trips["Rate_Code"] == "JFK") & (taxi_trips["Payment_Type"] == "Credit card") & (taxi_trips["Passenger_Count"] >=1) & (taxi_trips["Total_Amount"] < 90)]
JFK_Trips.head()
```
