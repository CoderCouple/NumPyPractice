## Key Components

- **Figure** — the outermost container (`plt.figure()`). Everything lives inside it. You can have multiple Axes per Figure.
- **Axes** — the actual subplot area where data is drawn (`fig.add_subplot()` or `plt.subplots()`). Not to be confused with the x/y axis lines — an Axes is the full region including ticks, labels, and the plot area.
- **Spines** — the four border lines of the Axes (top, bottom, left, right). You can hide or move individual spines, e.g. `ax.spines['top'].set_visible(False)`.
- **Artists** — everything rendered on the plot (lines, patches, text, collections) is an Artist. Lines, markers, bars, and scatter points all fall under this umbrella.

## Quick Reference

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Main title')

# axes[0,0] — line
x = np.linspace(0, 2*np.pi, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Line')

# axes[0,1] — bar
axes[0, 1].bar(['A','B','C','D'], [40, 80, 55, 90])
axes[0, 1].set_title('Bar')

# axes[1,0] — scatter
x2 = np.random.rand(50) * 100
axes[1, 0].scatter(x2, x2 * 0.8 + np.random.randn(50)*10)
axes[1, 0].set_title('Scatter')

# axes[1,1] — histogram
axes[1, 1].hist(np.random.randn(500), bins=20)
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
```





```python
fig, ax = plt.subplots()        # Figure + Axes
ax.plot(x, y, label='Line 1')   # Line artist + marker
ax.set_title('Chart title')
ax.set_xlabel('X axis label')
ax.set_ylabel('Y axis label')
ax.legend()
ax.grid(True)
```
