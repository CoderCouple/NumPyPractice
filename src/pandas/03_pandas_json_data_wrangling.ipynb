```python
import pandas as pd
```

```python
df1 = pd.read_json("../../data/pandas/example-1.json")
df2 = pd.read_json("../../data/pandas/example-2.json")
df_full = pd.concat([df1, df2])
df_full.reset_index()
```

```python
df_sales = pd.DataFrame(columns=['Date', 'Amount'])
df_sales
```

```python
new_data = {
    "Date": ["January 1", "January 2"],
    "Amount": [50000, 50000],
}
new_df_sales = pd.DataFrame(new_data)
new_df_sales
```

```python
df_sales_full = pd.concat([df_sales, new_df_sales])
df_sales_full
```

```python
df_second = pd.DataFrame({"Date":["January 3","January 4" ], "Amount":[5600,5400]})
df_second
```

```python
df_sls = pd.concat([df_sales_full, df_second])
df_sls
```

```python
df_sls.reset_index(drop=True, inplace=True)
df_sls
```
