```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../data/numpy/train.csv")
df.head()
```

```python
overall_quality = np.array(df["OverallQual"])[0:10]
overall_quality
```

```python
overall_condition = np.array(df["OverallCond"])[0:10]
overall_condition
```

```python
house_quality = overall_condition * overall_quality
house_quality
```
