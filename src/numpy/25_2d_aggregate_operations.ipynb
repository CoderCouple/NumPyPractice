```python
import numpy as np
import pandas as pd

df = pd.read_csv("../../data/numpy/train.csv")
df.head()
```

```python
sales_price =  np.array(df["SalePrice"])[0:10]
sales_price
```

# Sum

```python
sales_price.sum()
np.sum(sales_price)
```

# Mean (Average)

```python
sales_price.mean()
np.mean(sales_price)
```

# Standard Deviation


```python
sales_price.std()
np.std(sales_price)
```

# Minimum

```python
np.min(sales_price)
sales_price.min()
```

# Maximum

```python
np.max(sales_price)
sales_price.max()
```

# Argmin (Index of the minimum element)

```python
np.argmin(sales_price)
sales_price.argmin()
```

# Argmax (Index of the maximum element)

```python
np.argmin(sales_price)
sales_price.argmin()
```
