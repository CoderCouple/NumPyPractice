# Sequence to Array

```python
import numpy as np
```

```python
ints =  [-2,-1,0.,1,2]
arr1 = np.array(ints)
print(arr1)
```

```python
matrix = np.array([[1,2,3],
                   [4,5,6]]
                  )
print(matrix)
```

```python
matrix_3d = np.array([[[1,2,3]]])
print(matrix_3d)
```

```python
arr2 = np.array([1,2,3,-1], dtype=np.short)
print(arr2)
```
