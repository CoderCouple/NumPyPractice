```python
import numpy as np
```

```python
first_five = np.arange(5)
print(first_five)
```

```python
start_at_minus_2 = np.arange(-2,3)
print(start_at_minus_2)
```

```python
first_evens = np.arange(0,11, 2)
print(first_evens)
```

```python
first_negatives = np.arange(-1,-4, -1)
print(first_negatives)
```

```python
floats = np.arange(0.1,0.6, 0.1, dtype=np.single)
print(floats)
```

```python
ln = np.linspace(0.1, 0.5, 5)
print(ln)
```
