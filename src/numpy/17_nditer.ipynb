# nditer

```python
import numpy as np
```

```python
a = np.arange(16).reshape(4,4)
print(a)
```

```python
for element in a:
    print(element)
```

```python
# nditer

for element in np.nditer(a):
    print(element)
```

```python
# order parameter
for element in np.nditer(a, order='F'):
    print(element)
```

```python
# op_flags parameter - allows modifying while reading
for element in np.nditer(a, op_flags=['readwrite'], order='F'):
    element[...] = 0

print(a)
```

```python
# Modify 2 arrays at the same time
p = np.arange(12).reshape(4,3)
q = np.arange(12).reshape(4,3)

for x,y in np.nditer([p,q], op_flags=['readwrite']):
    print(x * y)
```
