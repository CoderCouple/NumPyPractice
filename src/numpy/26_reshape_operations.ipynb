```python
import numpy as np
import pandas as pd
from numpy.ma.core import arange

df = pd.read_csv("../../data/numpy/train.csv")
df.head()
```

# Reshaping

```python
sale_price_1d = np.array(df["SalePrice"])[0:12]
sale_price_1d
```

```python
sale_price_2d = sale_price_1d.reshape((4,3))
sale_price_2d
```

```python
sale_price_3d=sale_price_1d.reshape((3,2,2))
sale_price_3d
```

# Flattening

```python
sp_1d = sale_price_3d.flatten()
sp_1d
```
