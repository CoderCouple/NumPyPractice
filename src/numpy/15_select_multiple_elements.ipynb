# Advance Indexing

```python
import numpy as np
```

```python
arr2d = np.arange(16).reshape(4,4)
print(arr2d)
```

```python
print(arr2d[[1,3]])
```

```python
# Boolean Indexing
arr = np.arange(5)
print(arr)
```

```python
print(arr[arr >1])
```

```python
print(arr[np.array([True,False,True,False,False])])
```
