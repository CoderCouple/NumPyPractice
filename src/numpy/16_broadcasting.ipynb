# Broadcasting

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.broadcast_arrays(a, b)
print(c)
```

```python
d = a + b
print(d)
```

```python
e = a + b.T
print(e)
```

```python
p = np.arange(12).reshape(3, 4)
q = np.arange(12).reshape(3, 4)
print(p)
print(q)
```

```python
r = p + q.T
print(r)
```
