# Array Slicing

<img src="../../images/array_slicing.png" alt="Array Slicing" width="500">

* An important first distinction from Python's built-in lists is that array slices are views on the original array. This means that the data is not copied, and any modifications to the view will be reflected in the source array.
* The "bare" slice [:] will assign to all values in an array

```python
import numpy as np
a = np.arange(10)
print(a)
```

```python
b= np.arange(0,12).reshape(3,4)
print(b)
```

```python
a[:]
```

```python
b[0:1]
```

```python
b[1]
```

```python
b[1,0:2]
```

```python
b[1,0:2] = 5
b
```

## Array Indexing & Slicing Combined

<img src="../../images/array_indexing_slicing.png" alt="Array Indexing and Slicing" width="500">

```python
x = np.arange(1,10).reshape(3,3)
print(x)
```

```python
x[:2,1:]
```

```python
x[2]
x[2, :]
x[2:, :]
```

```python
x[:,:2 ]
```

```python
x[1:2,:2]
x[1,:2]
```
